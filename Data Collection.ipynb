```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
url = "https://www.espncricinfo.com/series/icc-men-s-t20-world-cup-2024-1411166/india-vs-south-africa-final-1415755/ball-by-ball-commentary"
```

```python
driver = webdriver.Chrome()
```

```python
driver.get(url)
```

```python
html = driver.page_source
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
ball = soup.find_all("div" , class_ = "ds-leading-none ds-mb-0.5")
for i in ball:
    print(i.text)
```

```python
ball1 = soup.find_all("div" , class_ = "ds-leading-none ds-mb-0.5")
for i in ball1:
    print(i.text)
```

```python
ballcount = soup.find_all("div", class_ = "lg:ds-flex lg:ds-items-center lg:ds-px-2")
for j in ballcount:
    print(j.text)
```

```python
ballcount1 = soup.find_all("div", class_ = "lg:ds-flex lg:ds-items-center lg:ds-px-2")
for j in ballcount1:
    print(j.text)
```

```python
c = soup.find_all("div", class_ = "ds-rounded ds-bg-fill-content-alternate ds-ml-4 lg:ds-ml-3 ds-mt-1 ds-inline-block ds-p-1")
for i in c:
    print(i.text)
```

```python
c1 = soup.find_all("div", class_ = "ds-rounded ds-bg-fill-content-alternate ds-ml-4 lg:ds-ml-3 ds-mt-1 ds-inline-block ds-p-1")
for i in c1:
    print(i.text)
```
